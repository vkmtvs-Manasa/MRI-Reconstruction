```python
import cv2
import pywt
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('test.avif', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Image not found!")
    exit()

print("Shape:", img.shape)

# Apply wavelet
coeffs2 = pywt.dwt2(img, 'haar')
LL, (LH, HL, HH) = coeffs2

# Show
plt.imshow(LL, cmap='gray')
plt.title("LL")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("Original")

plt.subplot(1,3,2)
blur = cv2.GaussianBlur(img, (15,15), 0)
plt.imshow(blur)
plt.title("Blur")

plt.subplot(1,3,3)
import numpy as np

small = img[::4, ::4]  # downsample
alias = np.repeat(np.repeat(small, 4, axis=0), 4, axis=1)
plt.imshow(alias)
plt.title("Aliased")

plt.show()
```
